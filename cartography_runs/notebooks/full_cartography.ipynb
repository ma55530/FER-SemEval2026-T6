# SemEval/QEvasion Dataset Cartography

This notebook inspects the outputs produced by `semeval_cartography.py`. It is intentionally output-first: run the remote Docker job, extract `output/`, then use these cells for reviewer-facing tables, plots, and qualitative examples.

```python
from pathlib import Path
import json
import pandas as pd
from IPython.display import display, Image, Markdown

BASE = Path.cwd()
OUT = BASE / "output"
RES = OUT / "results"
PLOTS = OUT / "plots"
EXAMPLES = OUT / "examples"

summary_path = RES / "summary_semeval_cartography.json"
summary = json.loads(summary_path.read_text(encoding="utf-8")) if summary_path.exists() else {}
summary
```

## Headline Tables

```python
for name in [
    "qevasion_constrained_region_counts.csv",
    "qevasion_constrained_region_by_clarity.csv",
    "qevasion_constrained_region_by_evasion.csv",
    "qevasion_constrained_hard_evasion_enrichment.csv",
    "qevasion_all_metrics.csv",
]:
    path = RES / name
    if path.exists():
        display(Markdown(f"### {name}"))
        display(pd.read_csv(path))
    else:
        print(f"Missing: {path}")
```

## Plots

```python
plot_names = [
    "qevasion_constrained_cartography.png",
    "qevasion_constrained_region_transitions.png",
    "qevasion_constrained_region_by_clarity.png",
    "qevasion_constrained_region_by_evasion.png",
    "qevasion_constrained_hard_evasion_enrichment.png",
    "qevasion_constrained_metadata_vs_difficulty.png",
    "qevasion_constrained_seed42_epoch_snapshots.png",
    "qevasion_constrained_seed42_confidence_trajectories.png",
    "qevasion_constrained_seed42_forgetfulness.png",
    "qevasion_constrained_test_seed42_confusion_clarity.png",
    "qevasion_constrained_test_seed42_confusion_evasion.png",
    "qevasion_constrained_test_seed42_confusion_pair.png",
]
for name in plot_names:
    path = PLOTS / name
    if path.exists():
        display(Markdown(f"### {name}"))
        display(Image(filename=str(path)))
    else:
        print(f"Missing: {path}")
```

## Qualitative Examples

```python
example_files = [
    "qevasion_constrained_examples_by_region.csv",
    "qevasion_constrained_hard_examples_by_evasion.csv",
    "qevasion_constrained_ambiguous_high_variability.csv",
    "qevasion_constrained_chronically_unstable.csv",
    "qevasion_constrained_test_high_confidence_wrong.csv",
    "qevasion_constrained_test_confusion_examples.csv",
]
for name in example_files:
    path = EXAMPLES / name
    if path.exists():
        display(Markdown(f"### {name}"))
        df = pd.read_csv(path)
        display(df.head(20))
    else:
        print(f"Missing: {path}")
```
